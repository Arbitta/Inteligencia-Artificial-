```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #from matplotlib import pyplot
```

```python
datos = pd.read_csv("../datos/WhiskyConFaltantes.csv") #levantamos el archivo a un dataframe
```

```python
datos
```

```python
datos['Precio'].fillna(datos['Precio'].mean(),inplace=True)
```

```python
categorias = datos['Categoria'].value_counts()
datos['Categoria'].replace('lujo','Lujo',inplace=True)
datos
```

```python
datos['Categoria']=datos['Categoria'].replace('lujo','Lujo')
```

```python
datos
```

```python
#datos['Anejamiento'].fillna(7)
datos['pb']=datos['Precio']/datos['Anejamiento']
datos
```

```python
etiq=['barato','medio','caro','muy caro']
columna = pd.cut(datos['Precio'],bins=len(etiq),labels=etiq)
datos['Precio2']=pd.Series.to_frame(columna)
datos
```

```python
from sklearn.preprocessing  import LabelEncoder

numerizacion = LabelEncoder()
datos['Categoria']=numerizacion.fit_transform(datos['Categoria'])
datos
```

```python
datos
```

```python
from sklearn.preprocessing import MinMaxScaler

normalizador= MinMaxScaler() #creamos el tipo de normalización

datosANormalizar=datos['Anejamiento'] #seleccionamos la o las columnas a normalizar

datosANormalizar=datosANormalizar.values.reshape(len(datosANormalizar),1) #en caso de que sea una sola hay que redimensionarla

datosNormalizados= normalizador.fit_transform(datosANormalizar) #se aplica el normalizador

datos['Anejamiento']=datosNormalizados #nuevamente se agrega al Dataframe
```
