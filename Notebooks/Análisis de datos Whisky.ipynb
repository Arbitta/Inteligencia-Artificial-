```python
label1=['precio']
label2=['malta']
fig, (ax1,ax2)=plt.subplots(nrows=1,ncols=2,figsize=(9,4))
plot1= ax1.boxplot(datos['Precio'],vert=True,patch_artist=True,labels=label1)
ax1.set_title('Precio del whisky')
plot2= ax2.boxplot(datos['Malta'],vert=True,patch_artist=True,labels=label2, notch=True)
ax2.set_title('Malta del whisky')
```

# Histograma

```python
plt.hist(datos['Precio'],bins=4) #bins=[55,81.25][81.25,107.5] [107.5,133.7] [133.7,160]
minimo=datos['Precio'].min()
maximo=datos['Precio'].max()
bins=6
rango = (maximo-minimo)/bins
rango
plt.xlabel('Precio')
plt.ylabel("Frecuencia")
plt.title("Histograma de precios del whisky")
```

# Diagrama de barras

```python
barras = datos['Categoria'].value_counts() #variables categoricas y devuelve las categorias
barras
```

```python
labels = np.arange(len(barras))
labels
```

```python
plt.bar(labels,barras)
plt.xticks(labels,barras.index,rotation=45)
```

# Diagrama de torta o gráfico de sectores

```python
plt.pie(barras,labels=barras.index)
```

# diagrama de dispersión o scatter

```python
plt.scatter(datos['Precio'],datos['Malta'],c={'red'})
plt.ylabel('Malta')
plt.xlabel('Precio')
```
